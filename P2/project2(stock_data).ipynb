## Set up the Alpaca API to be used

```python
# imports
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi

from datetime import date
from datetime import timedelta
from datetime import datetime
```

```python
# call Alpaca api

# Load .env enviroment variables
load_dotenv() 

# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

# Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

## Getting the tickers and data together

```python
#add SPX & VIX 
```

```python
tech_stocks = ["AAPL","MSFT","AMZN","FB","NVDA","GOOGL","KEYS","V","MA","CSCO","ADBE","AVGO","NFLX","INTC",
"CRM","ACN","QCOM","PYPL","TXN","AMD","INTU","AMAT","ORCL","SHOP","IBM","MU","NOW","LRCX","ADP","ADI",
"FIS","MRVL","KLAC","SNOW","FISV","NXPI","ADSK","SQ","TEL","ATVI","SNPS","XLNX","PANW","APH","WDAY",
"MCHP","CTSH","CDNS","HPQ","GPN","MSI","FTNT","SNAP","PAYX","EBAY","EA","CRWD","TWLO","DDOG","ZM",]

consumer_staple_stocks = ["PG","PEP","KO","PM","CVS","MO","MDLZ","CL","KMB","GIS","SYY","ADM","MCK","WBA","STZ","KR","CTVA","MNST",
"HSY","KDP","TSN","KHC","CHD","MKC","CLX","ABC","K","CAG","SJM","HRL","BG","IFF","DAR","LW","TAP","CPB",
"USFD","CASY","POST","INGR","FLO","ACI","HLF","SAM","SPB","BYND","FRPT","HAIN","FMX","GO","OLPX","REYN",
"PPC","SEB","CCEP","UL","DEO","EL","BTI","ABEV",]

utility_stocks = ["VST","WEC","SR","AGR","ES","NEP","WTRG","NOVA","OGS","CWEN","NWN","EVRG","DUK","AWK","POR","NEW","ORA","ALE",
"LNT","AVA","UTL","UGI","SWX","SO","EIX","SJI","SJW","PNM","PEG","PCG","PPL","EXC","OTTR","OGE","NI","NJR","NFG",
"XEL","MSEX","MGEE","MDU","CNP","YORW","HE","NEE","ETR","ATO","IDA","DTE"]

# these are not on alpaca
etfs = ["SPY"]
```

```python
# Format current date as ISO format

today = date(2021,1,21)
ending_date = date(2018,8,21)

# Set the tickers
tickers = tech_stocks + consumer_staple_stocks + utility_stocks + etfs


# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices
# (use a limit=1000 parameter to call the most recent 1000 days of data)

df_investments = alpaca.get_barset(
    tickers,
    timeframe,
    start = today,
    end= ending_date,
    limit=1000
).df

#Drop all NaNs, but at what cost?
df_investments.dropna(inplace=True)

# Preview DataFrame
df_investments.head(10)
```

```python
# turn data into actual dataframe to leave only closing prices

# Create a DataFrame filtering only the index and Close columns

idx = pd.IndexSlice
df_investments = df_investments.loc[:,idx[:,'close']]
# Review the DataFrame
df_investments.head(20)
```
